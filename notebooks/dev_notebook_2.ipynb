```python
import pandas as pd
import numpy as np
import duckdb
import json
```

```python
conn = duckdb.connect('../data/bicimad.db',read_only=True)
conn
```

```python
type(conn)
```

```python
my_sql = conn.sql('SELECT* FROM bicimad_stations')
my_sql
```

```python
df = my_sql.df()
df
```

```python
df['geometry.coordinates']
```

```python
bicimad = df[['name','address','geometry.coordinates']]
bicimad
```

```python
bicimad.columns = ['BiciMAD Station','Station Location','Coordinates']
bicimad
#type(bicimad)
```

```python
bicimad['Coordinates'] = bicimad['Coordinates'].str.strip('[]')
bicimad
```

```python
bicimad[['Latitude', 'Longitude']] = bicimad['Coordinates'].apply(lambda x: pd.Series({'Latitude': x[0], 'Longitude': x[1]}))
bicimad
```

```python
bicimad[['Longitude', 'Latitude']] = bicimad['Coordinates'].str.split(',', expand=True)
bicimad
```

```python
bicimad.drop(['Coordinates'], axis=1)
```

```python
bicimad.to_csv('../data/bicimad.csv', index=False)
```
